# Drop unnamed column

```python
df = df.drop("Unnamed: 0", axis = 1)
print(df.head())
```

# Gender Distribution

```python
plt.figure(figsize = (5,5))
ax = sns.countplot(data = df, x = "Gender")
ax.bar_label(ax.containers[0])
plt.title("Gender Distribution")
plt.show()
```

No. of female students are gretaer than no. of male students

```python
gb = df.groupby("ParentEduc").agg({"MathScore":'mean',"ReadingScore":'mean',"WritingScore":'mean'})
print(gb)
```

```python
plt.figure(figsize = (4,4))
sns.heatmap(gb, annot = True)
plt.title("Relationship between Parent's Education and Student's Score")
plt.show()
```

from the above chart we can conclude that parents education have good impact on their students marks

```python
gb1 = df.groupby("ParentMaritalStatus").agg({"MathScore":'mean',"ReadingScore":'mean',"WritingScore":'mean'})
print(gb1)
```

```python
plt.figure(figsize = (4,4))
sns.heatmap(gb1, annot = True)
plt.title("Relationship between Parent's Marital Status and Student's Score")
plt.show()
```

From the above chart we can conclude that there is negligible impact on student's score due to their parents marital status

# Distribution of Ethinic Group

```python
groupA = df.loc[(df["EthnicGroup"] == "group A")].count()
groupB = df.loc[(df["EthnicGroup"] == "group B")].count()
groupC = df.loc[(df["EthnicGroup"] == "group C")].count()
groupD = df.loc[(df["EthnicGroup"] == "group D")].count()
groupE = df.loc[(df["EthnicGroup"] == "group E")].count()

l = ["Group A", "Group B","Group C","Group D","Group E"]
my_list = [groupA["EthnicGroup"],groupB["EthnicGroup"],groupC["EthnicGroup"],groupD["EthnicGroup"],groupE["EthnicGroup"]]

plt.pie(my_list,  labels = l, autopct = "%1.2f%%" )
plt.title("Distribution of Ethnic Groups")
plt.show()
```

# Weekly Study Hours

```python
gb2 = df.groupby("WklyStudyHours").agg({"MathScore":'mean',"ReadingScore":'mean',"WritingScore":'mean'})
print(gb2)
```

```python
plt.figure(figsize = (4,4))
sns.heatmap(gb2, annot = True)
plt.title("Relationship between Weekly Study Hours and Student's Score")
plt.show()
```

No. of Weekly Study Hours impacts Student's scores at a certain level.
